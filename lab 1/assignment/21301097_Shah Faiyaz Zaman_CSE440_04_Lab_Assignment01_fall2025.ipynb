import sttop words for removal

***Note: the current format of word/pos tag isn't proccessable by wordnet lemmatizer***
Futhermore, wornetlemmattizer only acceptts params of pos in given format:

Lemmatize word using WordNets built-in morphy function. Returns the input word unchanged if it cannot be found in WordNet.

Parameters
word (str)  The input word to lemmatize.

pos (str)  The Part Of Speech tag. Valid options are n for nouns, v for verbs, a for adjectives, r for adverbs and s for satellite adjectives.

pos  str

Returns
The lemma of word, for the given pos.

Return type
str

So, VBG and otthers need to be converted to pos tags that is readable by lemmatizer


```python

def get_wordnet_pos(tag):
    if tag.startswith('j'):
        return "a"
    elif tag.startswith('v'):
        return 'v'
    elif tag.startswith('n'):
        return 'n'
    elif tag.startswith('r'):
        return 'r'
    else:
        return 'n'
```

```python
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
filtered_tokens=[]
readable_pos_tags=[]
for token in fiction_word_tokens:
    
    token_temp=token.split("/")
    if token.lower() not in stop_words:
        filtered_tokens.append(token_temp[0])
       
        if len(token_temp)>1:
            pos_tag_temp=get_wordnet_pos(token_temp[1]) 
            readable_pos_tags.append(pos_tag_temp)
        else:
            readable_pos_tags.append('n')  # Default to noun if no POS tag is present
            
# print(filtered_tokens[:100])
# for j in filtered_tokens:

#     j.split("/")

print("after removing stop words:", filtered_tokens[:100])
print("POS tags for the filtered tokens:", readable_pos_tags[:100])
```

apply lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = []
for token in  range (len(filtered_tokens)):
    lemmatized_token = lemmatizer.lemmatize(filtered_tokens[token], pos=readable_pos_tags[token])
    lemmatized_tokens.append(lemmatized_token)
print(lemmatized_tokens[:100])
```

```python
#ttest lemmattized orr nott
for i in range(10):
    print(f"Original token: {filtered_tokens[i]} --> Lemmatized token: {lemmatized_tokens[i]}")
```

```python
from nltk.probability import FreqDist
fd3 = FreqDist(lemmatized_tokens)
fd3.most_common(15)
word = []
count = []
for w, c in fd3.most_common(15):
    word.append(w)
    count.append(c)
print(word)
print(count)
plt.bar(word, count)
```

```python
nltk.download('gutenberg')
from nltk.corpus import gutenberg
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
for i in dir(gutenberg):
    print(i)
print(gutenberg.fileids())
hamlet = gutenberg.raw('shakespeare-hamlet.txt')
```

preprocess: to lower-> punc removal->stopwords removal

```python
lower_cased = hamlet.lower()
word_tokens = word_tokenize(lower_cased)
print(word_tokens[:100])
punctuattions=['.', ',', '!', '?', ';', ':', '-', '_', '(', ')', '[', ']', '{', '}', '"', "'", '...']
for word in word_tokens:
    if word in punctuattions:
        word_tokens.remove(word)
print("after punc removal:", word_tokens[:100])
stop_words = stopwords.words('english')
filtered_tokens_hamlet=[]
for word in word_tokens:
    if word not in stop_words:
        filtered_tokens_hamlet.append(word)
print("after stopwords removal:", filtered_tokens_hamlet[:100])
```

20 most freq word

```python
from nltk.probability import FreqDist
frequenncy_dist = FreqDist(filtered_tokens_hamlet)
most_common_20 = frequenncy_dist.most_common(20)
```

Bi grams

```python
from nltk.util import ngrams
bigrams = list(ngrams(filtered_tokens_hamlet, 2))
print(bigrams[:20])
   
```

co occurence matrix from bigram

```python
import numpy as np
co_occurrence_matrix = np.zeros((20, 20), dtype=int)
word_to_index={}
for i,word in enumerate(most_common_20):
    word_to_index[word[0]]=i  
print(word_to_index)   
for word1,word2 in bigrams:
    if word1 in word_to_index and word2 in word_to_index:
        index1 = word_to_index[word1]
        index2 = word_to_index[word2]
        co_occurrence_matrix[index1][index2] += 1
print("Co-occurrence Matrix:")
print(co_occurrence_matrix)
```

```python
nltk.download('udhr')
from nltk.corpus import udhr
from nltk.tokenize import word_tokenize
for j in udhr.fileids():
    if j.startswith('Hungarian'):
        print(j)
hungarian_text = udhr.raw('Hungarian_Magyar-UTF8')
print(hungarian_text[:1000])
```

```python
vowels=['a', 'e', 'i', 'o', 'u'] #ccannot type accented vowels in english keyboard so only english vowels are used
word_token=word_tokenize(hungarian_text)
vower_sequence=[]
for word in word_token:
    vowels_only=""
    for char in word.lower():
        if char in vowels:
            vowels_only+=char
    if vowels_only:
        vower_sequence.append(vowels_only)
print(vower_sequence[:100])
```

```python
from nltk.util import ngrams
from nltk.probability import FreqDist
triigrams = list(ngrams(vower_sequence, 3))
print(triigrams[:20])
trigram_freq_dist = FreqDist(triigrams)
most_common_trigrams = trigram_freq_dist.most_common()
print(most_common_trigrams)
sorted_trigrams = sorted(most_common_trigrams, key=lambda x: x[1],)
for k,v in sorted_trigrams:
    print(f"Trigram: {k}, Frequency: {v}")

```
