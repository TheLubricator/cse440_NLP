**Q2**

```python
from nltk.corpus import gutenberg
from nltk.tokenize import word_tokenize
nltk.download('gutenberg')
from wordcloud import WordCloud
nltk.download('punkt_tab') # Download the punkt tokenizer models
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
```

```python
gutenberg.fileids()
```

```python
austen_emma = gutenberg.raw('austen-emma.txt')
austen_persuasion = gutenberg.raw('austen-persuasion.txt')
austen_emma_word_tokens = word_tokenize(austen_emma)
austen_persuasion_word_tokens = word_tokenize(austen_persuasion)
# print("austen emma text example words:", austen_emma[:1000])
# print("example tokens from austen-emma.txt:", austen_emma_word_tokens[:1000])
```

```python
fd1 = FreqDist(austen_emma_word_tokens)

wc2 = WordCloud(width = 1000, height = 600, background_color= 'white')
cloud1 = wc2.generate_from_frequencies(fd1)
plt.figure(figsize = (10,6))
plt.imshow(cloud1)
plt.axis('off')
```

```python
fd2 = FreqDist(austen_persuasion_word_tokens)

wc2 = WordCloud(width = 1000, height = 600, background_color= 'white')
cloud2 = wc2.generate_from_frequencies(fd1)
plt.figure(figsize = (10,6))
plt.imshow(cloud2)
plt.axis('off')
```
