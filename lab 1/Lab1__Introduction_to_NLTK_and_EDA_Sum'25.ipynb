# Stemming and Lemmatization

```python
from nltk.stem import LancasterStemmer, PorterStemmer, SnowballStemmer
```

```python
words = ['argue','argued','argument','arguing','arguments']

stemmer = PorterStemmer()

stemmed_word = []
for word in words:
  stemmed_word.append(stemmer.stem(word))

print(stemmed_word)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemmatized_word = []
for word in words:
  lemmatized_word.append(lemmatizer.lemmatize(word, pos = 'v'))

print(lemmatized_word)
```

```python
# Comparative analysis
from tabulate import tabulate

def stemming_example(words):
    porter = PorterStemmer()
    lancaster = LancasterStemmer()
    snowball = SnowballStemmer("english")

    data = [[word, porter.stem(word), lancaster.stem(word), snowball.stem(word)] for word in words]
    headers = ["Word", "PorterStemmer", "LancasterStemmer", "SnowballStemmer"]
    print("Stemming Results:")
    print(tabulate(data, headers=headers, tablefmt="grid"))
    print("\n")

def lemmatization_example(words):
    lemmatizer = WordNetLemmatizer()

    data = [[word, lemmatizer.lemmatize(word, pos='n'), lemmatizer.lemmatize(word, pos='v')] for word in words]
    headers = ["Word", "Lemmatized (noun)", "Lemmatized (verb)"]
    print("Lemmatization Results:")
    print(tabulate(data, headers=headers, tablefmt="grid"))
    print("\n")

words = ["connect", "connected", "connecting", "connection", "connections",
         "argue", "argued", "arguing", "argument", "arguments",
         "nation", "national", "nationality", "nationalism", "nationalist",
         "act", "acted", "acting", "action", "active", "activity",
         "derive", "derived", "deriving", "derivative", "derivation"]

print("Original Words:", words)
print("\n")

stemming_example(words)
lemmatization_example(words)
```

# Co occurance Matrix

```python
list1 = ['a','b','c']
for i, word in enumerate(list1):
  print(i,word)
```

```python
import numpy as np

text = "Someone has told you that words in English can be separated by separated separated separated separated separated separated separated separated separated  simply splitting on whitespace. How many times would that heuristic fail for the following text?"
tokens = word_tokenize(text.lower())

unique_words = list(set(tokens))

#Initialize co-occurrence matrix (square matrix with zeros)
matrix = np.zeros((len(unique_words), len(unique_words)), dtype=int)

word_to_index = {}  #Create an empty dictionary

for i, word in enumerate(unique_words):
    word_to_index[word] = i  #Assigning index to the word

print(word_to_index)

for i in range(len(tokens)-1):
  current_word = tokens[i]
  next_word = tokens[i+1]
  cw_index = word_to_index[current_word]
  nw_index = word_to_index[next_word]
  matrix[cw_index,nw_index] += 1

print(matrix)
```

```python
from wordcloud import WordCloud

text = '''Very vinous - Burgundy with bubbles, to some extent and so a bit much in some ways.  I suspect it needs time to mellow - it will be interesting to see how this develops (this bottle was disgorged only last year).   If all goes well ***(*)
All quite refined and delicate, appley pinot fruit, autolytic flavours. Another fairly recent disgorgement after 4 years in cellars (credit the producer for an excellent back label), it just seems a bit lacking in intensity despite the obvious quality. Perhaps this will develop but for now a good ***
Good fruit - nice acidity. Pretty intense although it seems quite evolved for a young wine - I fear it may go down the Ozzy route of rather early flabby kerosene flavours.  Top  **
Curiously Sauvignon-like - quite rich. Odd balance - people were saying this is a vineyard Huet took on and then abandoned. Good wine but it doesn't seem to have the class of their major wines. A low **(*)
Pretty nice.   Clean, rich, yet decently taught.  A credit to the appelation.'''

tokens = word_tokenize(text)
fd = FreqDist(tokens)

wc = WordCloud(width = 1000, height = 600, background_color= 'white')
cloud = wc.generate_from_frequencies(fd)
plt.figure(figsize = (10,6))
plt.imshow(cloud)
plt.axis('off')
```
